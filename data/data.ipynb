```python
# Load the spectrogram data from the .npz file
spectrogram_data_real = np.load('./real.npz')
print(np.shape(spectrogram_data_real))
print((spectrogram_data_real))

for key in spectrogram_data_real.keys():
    print(key)  # This will print out the keys available in the .npz file


# Access the 'x' and 'y' arrays
x = spectrogram_data_real['x']  
y = spectrogram_data_real['y']  
print(np.shape(x))
print(np.shape(y))



# print(spectrogram_data_real['x'][0])

print(spectrogram_data_real['y'][0])
print(spectrogram_data_real['y'][1])
# print(spectrogram_data_real['y'][2])

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from scipy.signal import istft

# Assuming you have your spectrogram data in a variable named 'spectrogram'
# spectrogram.shape should be (1998, 101)

# Reconstruct the complex valued STFT data
_, audio = istft(spectrogram_data_real['x'][0], fs=44100)  # Adjust fs according to your original sampling rate

# Normalize the audio to be between -1 and 1
audio /= np.max(np.abs(audio))

# Write the audio to a WAV file
write("reconstructed_audio_real.wav", 44100, audio.astype(np.float32))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from scipy.signal import istft

# Assuming you have your spectrogram data in a variable named 'spectrogram'
# spectrogram.shape should be (1998, 101)

# Reconstruct the complex valued STFT data
_, audio = istft(spectrogram_data_fake['x'][0], fs=44100)  # Adjust fs according to your original sampling rate

# Normalize the audio to be between -1 and 1
audio /= np.max(np.abs(audio))

# Write the audio to a WAV file
write("reconstructed_audio_fake.wav", 44100, audio.astype(np.float32))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have your spectrogram data in a variable named 'spectrogram_data_fake'
# spectrogram_data_fake['x'] is assumed to be a list of spectrograms, each of shape (1998, 101)

plt.figure(figsize=(15, 10))  # Adjust the figure size as needed

num_spectrograms =5
for i, spectrogram in enumerate(spectrogram_data_fake['x'][0:num_spectrograms]):
    plt.subplot(num_spectrograms, 1, i+1)
    plt.imshow(np.abs(spectrogram.T), aspect='auto', origin='lower', cmap='viridis')
    plt.xlabel('Time')
    plt.ylabel('Frequency Bin')
    plt.title(f'Spectrogram {i+1} [Fake]')
    plt.colorbar(label='Magnitude')

plt.tight_layout()
plt.show()



for i, _ in enumerate(spectrogram_data_fake['x'][0:num_spectrograms]):
    print(spectrogram_data_fake['y'][i])
```
